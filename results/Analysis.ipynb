```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
stats = pd.read_csv('example_stats.csv')
```

```python
stats
```

```python
plt.bar(0.2, stats['Median Response Time'][0], 0.2, alpha = 0.8, color = 'brown', label='web app')
plt.bar(0.4, stats['Median Response Time'][1], 0.2, alpha = 0.8, color = 'maroon', label='tensorflow')
plt.bar(0.6, stats['Median Response Time'][2], 0.2, alpha = 0.8, color = 'grey', label='average')
plt.ylabel('Median Response Time [s]')
plt.title('Comparison of Performance - Median')
plt.legend()
plt.xticks([])
plt.show()
```

```python
plt.bar(0.2, stats['Average Response Time'][0], 0.2, alpha = 0.8, color = 'brown', label='web app')
plt.bar(0.4, stats['Average Response Time'][1], 0.2, alpha = 0.8, color = 'maroon', label='tensorflow')
plt.bar(0.6, stats['Average Response Time'][2], 0.2, alpha = 0.8, color = 'grey', label='average')
plt.ylabel('Average Response Time [s]')
plt.title('Comparison of Performance - Average Time')
plt.legend()
plt.xticks([])
plt.show()
```

```python
plt.bar(0.2, stats['Min Response Time'][0], 0.2, alpha = 0.5, color = 'blue', label='web app')
plt.bar(0.2, stats['Max Response Time'][0], 0.2, alpha = 0.5, color = 'blue')
plt.bar(0.4, stats['Min Response Time'][1], 0.2, alpha = 0.5, color = 'grey')
plt.bar(0.4, stats['Max Response Time'][1], 0.2, alpha = 0.5, color = 'grey', label='tensorflow')
plt.yscale('log')
plt.ylabel('Average Response Time [s] ')
plt.title('Comparison of Performance - Min-Max Time')
plt.legend()
plt.xticks([])
plt.show()
```
